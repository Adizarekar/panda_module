```python
def convert_people_cell(cell):
    if cell== "n.a.":
        return "Sam Walton"
    else:
        return cell
    
def convert_eps_cell(cell):
    if cell== "not available":
        return None
    else:
        return cell
    
df = pd.read_excel('F:\Education\DS_Roadmap\panda_module\stock_data.xlsx', 'Sheet1', converters={
    'people': convert_people_cell,
    'eps': convert_eps_cell
    
})

df
```

```python
df.to_excel('new.xlsx', sheet_name='stocks')
```
